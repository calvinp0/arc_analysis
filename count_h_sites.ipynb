```python

# Delete duplicates based on rxn	r2h	r1	r2	r1h	ts
coord_data_deduped = coord_data.drop_duplicates(subset=["rxn", "r2h", "r1", "r2", "r1h", "ts"])
coord_data_deduped = coord_data_deduped.reset_index(drop=True)
coord_data_deduped.head()


```

```python
# I need to take 'r1' and 'r1h' and check convert them to xyz - then select symbols from the dictionaries and make sure that r1's len is one less than r1h's len

def get_xyz_from_dict(r1, r1h):
    r1_xyz = check_xyz_dict("\n".join(literal_eval(r1)))
    r1h_xyz = check_xyz_dict("\n".join(literal_eval(r1h)))
    r1_symbols = r1_xyz['symbols']
    r1h_symbols = r1h_xyz['symbols']
    return r1_symbols, r1h_symbols

def check_len(r1, r1h):
    r1_symbols, r1h_symbols = get_xyz_from_dict(r1, r1h)
    if len(r1_symbols) == len(r1h_symbols) - 1:
        return True
    else:
        return False
    
coord_data['len_check_r1'] = coord_data.apply(lambda x: check_len(x['r1'], x['r1h']), axis=1)
coord_data['len_check_r2'] = coord_data.apply(lambda x: check_len(x['r2'], x['r2h']), axis=1)
```

```python
def convert_species(g_xyz, index_h, index_a, distances: list = [0.8, 1.0, 1.1]):
    """
    Try to pull atoms closer using specified distances. If none of them work, return g_xyz.

    Args:
        g_xyz (dict): The geometry in XYZ format.
        index_h (int): Index of the hydrogen atom to adjust.
        index_a (int): Index of the atom to pull closer.
        distances (list): List of distances to try pulling the atoms closer.

    Returns:
        dict: The modified geometry or the original geometry if no modification succeeded.
    """
    for i in distances:
        try:
            temp_xyz = h_abs_atoms.pull_atoms_closer(g_xyz, index_h, index_a, i)
            temp_mol = ARCSpecies(label="temp", xyz=temp_xyz)
        except:
            continue
        else:
            if i != 0.8:
                print("Utilized distance: ", i)
            return temp_xyz

    # If none of the distances worked, return g_xyz
    return g_xyz
```

```python
coord_data['g1_xyz'] = None
coord_data['g1_xyz'] = coord_data['g1_xyz'].astype('object')
coord_data['g2_xyz'] = None
coord_data['g2_xyz'] = coord_data['g2_xyz'].astype('object')
coord_data['index_dict'] = None
coord_data['index_dict'] = coord_data['index_dict'].astype('object')
connections = {}
errors = {}
for idx, row in coord_data.iterrows():
    g1_xyz = None
    g2_xyz = None

    index_dict = None
    try:
        ts_list = row['ts_lists']
        ts_str = "\n".join(ts_list)
        ts_xyz = check_xyz_dict(ts_str)
        ts_df = h_abs_atoms.convert_xyz_to_df(ts_xyz)
        ts_results = h_abs_atoms.get_h_abs_atoms(ts_df)

        # Split the data
        g1_xyz, g2_xyz, index_dict = split.get_group_xyzs_and_key_indices_from_ts(
            xyz = ts_xyz, a=ts_results['A'], b=ts_results['B'], h=ts_results['H']
        )

        g1_xyz_pulled = convert_species(g1_xyz, index_dict['g1_h'], index_dict['g1_a'])
        g2_xyz_pulled = convert_species(g2_xyz, index_dict['g2_h'], index_dict['g2_a'])

        claimed_alignments = set()  # To track already matched columns
        column_list_names = ['r1h', 'r2h']
        # Now we need to do a perception check to make sure the guessed molecules align with the original
        for j in [g1_xyz_pulled, g2_xyz_pulled]:
            for i, molecules in enumerate([row['r1h'], row['r2h']]):
                # Skip this column if it is already aligned
                if column_list_names[i] in claimed_alignments:
                    continue

                # Check for isomorphism alignment
                if perception.check_isomorphism(j, check_xyz_dict("\n".join(literal_eval(molecules)))):
                    # Add the matched column to the connections dictionary
                    if row['rxn'] not in connections:
                        connections[row['rxn']] = {column_list_names[i]: j}
                    else:
                        connections[row['rxn']][column_list_names[i]] = j

                    # Mark this column as claimed
                    claimed_alignments.add(column_list_names[i])
                    break
        if len(list(connections[row['rxn']].keys())) != 2:
            print("Could not find alignment for rxn: ", row['rxn'])
            continue


        # Add the data to the dataframe
        coord_data.at[idx, 'g1_xyz'] = json.dumps(g1_xyz_pulled)
        coord_data.at[idx, 'g2_xyz'] = json.dumps(g2_xyz_pulled)
        coord_data.at[idx, 'index_dict'] = json.dumps(index_dict)
    except Exception as e:
        print("Error: ", e)
        print("Could not process rxn: ", row['rxn'])
        errors[row['rxn']] = e
        continue
```
